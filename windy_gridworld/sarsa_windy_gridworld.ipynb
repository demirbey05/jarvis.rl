```python
from env import WindyEnv
import numpy as np
from tqdm import tqdm
import random
```

```python
class SarsaAgent:
    def __init__(self, env, alpha=0.5, epsilon=0.1, gamma=1):
        self.env = env
        self.alpha = alpha
        self.epsilon = epsilon
        self.gamma = gamma
        self.q = np.zeros((len(env.state_space), len(env.action_space)))

        self.train()
    
    def epsilon_greedy(self, s):
        if np.random.uniform() < self.epsilon:
            return random.choice(range(len(self.env.action_space)))
        else:
            return np.argmax(self.q[s])
    

    def train(self, episodes=1000000):
        for _ in tqdm(range(episodes)):
            state = self.env.reset()
            s = self.env.state_space.index(state)
            a = self.epsilon_greedy(s)
            
            i = 0
            while True:
                action = self.env.action_space[a]
                state_, r, done = self.env.step(action)
                s_ = self.env.state_space.index(state_)
                a_ = self.epsilon_greedy(s_)
                self.q[s, a] += self.alpha * (r + self.gamma * self.q[s_, a_] - self.q[s, a])
                s = s_
                a = a_
                if done:
                    i=0
                    break
                    
                i += 1
            

    
```

```python
agent = SarsaAgent(WindyEnv(10,7,[0,0,0,1,1,1,2,2,1,0]),alpha=0.1,epsilon=0.1,gamma=1)
```
