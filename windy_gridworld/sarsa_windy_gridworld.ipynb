```python
from env import WindyEnv
import numpy as np
from tqdm import tqdm
```

```python
class SarsaAgent:
    def __init__(self, env, alpha=0.5, epsilon=0.1, gamma=1):
        self.env = env
        self.alpha = alpha
        self.epsilon = epsilon
        self.gamma = gamma
        self.q = np.zeros((len(env.state_space), len(env.action_space)))
        self.pi = np.zeros((len(env.state_space), len(env.action_space)))
        self.pi[:, :] = 1 / len(env.action_space)
        print(self.pi)

        self.train()
    
    def derive_policy_from_q(self):
        self.pi = np.zeros((len(self.env.state_space), len(self.env.action_space)))
        for i in range(len(self.env.state_space)):
            best_a = np.argmax(self.q[i])
            self.pi[i, best_a] = 1 - self.epsilon + self.epsilon / len(self.env.action_space)
            for a in range(len(self.env.action_space)):
                if a != best_a:
                    self.pi[i, a] = self.epsilon / len(self.env.action_space)
    

    def train(self, episodes=1000):
        for _ in tqdm(range(episodes)):
            state = self.env.reset()
            s = self.env.state_space.index(state)
            a = np.random.choice([0,1,2,3], p=self.pi[s])
            
            i = 0
            while True:
                action = self.env.action_space[a]
                state_, r, done = self.env.step(action)
                s_ = self.env.state_space.index(state_)
                a_ = np.random.choice([0,1,2,3], p=self.pi[s_])
                self.q[s, a] += self.alpha * (r + self.gamma * self.q[s_, a_] - self.q[s, a])
                s = s_
                a = a_
                if done:
                    print(i)
                    i=0
                    break
                    
                i += 1
            

    
```
